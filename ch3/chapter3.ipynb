## Creating synthetic images with OpenCV

In the following program, we will start working with arrays, and introduce a simple and fast way of creating images, in this case a white 500x500 array,

```python
//numpy_array.py
import numpy as np
import cv2

img = np.zeros((500,500), dtype=np.float)
img[0:500, 0:500] = 1

cv2.imshow('image',img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

Here is the image produced by the code. I have added a border added for definition purposes, so it can be more easily seen by the reader.

```python
from IPython.display import Image
Image(filename='/home/cesco/Desktop/book_entirety/new_chapter_3/images/B04209_03_01.png')
```

As you can see in this caption, the white pixels have been extended to include all of the alloted space in the array created by the np.zeros() function.  You can manipulate pixels in the image through the slicing of the array, and this way you can reduce the size of the white portion of the image from a 500x500  to a 200x200  portion with just a couple lines of code in Python,

```python
//numpy_array3.py

import numpy as np
import cv2



img = np.zeros((500,500), dtype=np.float)
img[150:350, 150:350] = 1

cv2.imshow('image',img)

cv2.waitKey(0)
cv2.destroyAllWindows()
```

In the following program we continue to integrate some of  the tools of the Python ecosystem with OpenCV.  We also introduce  matplotlib, a graphing tool, and explore a new way of creating random images. This randomness will come to play later as we find ways to add random noise to images, as we test different filter and image processing techniques.

The first three lines of the program import the modules to use in our program, and use the np.zeros, and np.random to generate arrays. The last lines of code direct the program to plot the two images side by side, and equally balanced.

** Importing NumPy as np and PyPlot as plt is done by convention in order to reduce the amount of typing into our programs.

```python
//numpy_mask2.py
 
import numpy as np
import cv2
from matplotlib import pyplot as plt

img2 = np.zeros((500,500), np.int8)
random = np.random.rand(500, 500)

plt.subplot(131),plt.imshow(img2,'gray'),plt.title('Mask-0')
plt.xticks([]),plt.yticks([])

plt.subplot(133),plt.imshow(random,'gray'),plt.title('Random')
plt.xticks([]),plt.yticks([])

plt.show()
```

When we execute the code we get two subplots, the first one is an all black  image, product of using the np.zeros () to create a synthetic black image, and the second one is an agglomeration of random black, white, and gray blotches, as shown here,

We can instead import it from a file, a video, a mobile device, or grabbed directly from an Internet capture,  so we can do some array manipulations to it. **The development and maintenance of the Python Image Library, PIL has slowed down since its last release in 2009,  but there is an actively-developed fork of PIL called Pillow.

You can import , and display an image using OpenCV, with the following,

```python
//cv2_display.py
#!/usr/bin/env python 

import cv2

original = cv2.imread('home_tree.png')
gray = cv2.imread('home_tree.png', 0)

cv2.imshow('Original', original)
cv2.imshow('Gray', gray)

cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
//matplot_pil.py

from matplotlib import pyplot as plt
import matplotlib.image as mpimg
import cv2

original = mpimg.imread('home_tree.png')
gray = cv2.imread('home_tree.png',0)

plt.subplot(131),plt.imshow(gray, cmap = 'gray'),plt.title('Grayscale Image')
plt.xticks([]), plt.yticks([])

plt.subplot(133),plt.imshow(original, cmap='gray'),plt.title('Original Image')
plt.xticks([]), plt.yticks([])

plt.show()
```

## A closer look at the N-dimensional array

In this section we will take a closer look at the NumPy N-dimensional arrays.  An N-dimensional array, or ndarray as it is most commonly known is a multi-dimensional container of items of the same type and size. The N  n is for the number of dimensions number and items in an array are defined by its shape, and the type of items is specified by a data-type object, most commonly known as a dtype.  
In the following code, this little snippet will read the image with cv2, print the dtype of the image, and size of the image,

```python
//cv2file.py
import cv2 

img = cv2.imread('beach.png') 

print type(img) 
print 'img shape: ', img.shape 
print 'img.dtype: ', img.dtype 
print 'img.size: ', img.size 

cv2.imshow('Original', img)
cv2.waitKey(0) 
cv2.destroyAllWindows() 
```
